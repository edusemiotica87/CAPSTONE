SpaceX has gained worldwide attention for a series of historic milestones.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file from the specified path
file_path = 'C:\\Users\\dedof\\Downloads\\Spacex.csv'
spacex_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(spacex_df.head())

# Visualize the relationship between different parameters

# Scatter plot between Payload Mass (kg) and Launch Outcome
plt.figure(figsize=(15, 10))
sns.scatterplot(data=spacex_df, x='Payload Mass (kg)', y='Outcome', hue='Orbit')
plt.title('Payload Mass (kg) vs. Outcome by Orbit')
plt.show()

# Scatter plot between Launch Site and Launch Outcome
sns.catplot(data=spacex_df, x='Launch Site', hue='Outcome', kind='count', height=5, aspect=2)
plt.title('Launch Site vs. Outcome')
plt.show()

# Scatter plot between Booster Version and Launch Outcome
sns.catplot(data=spacex_df, x='Booster Version Category', hue='Outcome', kind='count', height=5, aspect=2)
plt.title('Booster Version Category vs. Outcome')
plt.show()

# Visualize the launch success yearly trend

# Extracting the year from the Date column
spacex_df['Date'] = pd.to_datetime(spacex_df['Date'])
spacex_df['Year'] = spacex_df['Date'].dt.year

# Plotting the trend of launch success over the years
plt.figure(figsize=(15, 5))
sns.countplot(data=spacex_df, x='Year', hue='Outcome')
plt.title('Launch Success Yearly Trend')
plt.show()

# Create dummy variables for categorical columns

# Creating dummy variables
spacex_df_dummies = pd.get_dummies(spacex_df, columns=['Launch Site', 'Orbit', 'Booster Version Category'], drop_first=True)

# Display the first few rows of the new dataframe with dummy variables
print(spacex_df_dummies.head())
```
